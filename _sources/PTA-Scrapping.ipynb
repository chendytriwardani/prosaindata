```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
listPTA = []

for i in range(152,173):
  URL = "https://pta.trunojoyo.ac.id/c_search/byprod/10/" + str(i+1)
  req = requests.get(URL)
  soup = BeautifulSoup(req.text, 'html.parser')
  items = soup.find('div', {'id':'content_journal'})

  for it in items.findAll('li',{'data-cat':'#luxury'}):
    tempNama = it.find('span').text.strip('\n').replace('Penulis : ', '')
    link = it.find('a', 'gray button')['href']
    req2 = requests.get(link)
    soup2 = BeautifulSoup(req2.text, 'html.parser')
    abs = soup2.findAll('div')
    abstrak = soup2.find('p', {'align':'justify'}).text
  
    tempNIM = link.replace('https://pta.trunojoyo.ac.id/welcome/detail/', '')

    listPTA.append([it.a.text.strip('\n'), tempNIM, tempNama, abstrak])
```

```python
listPTA_df = pd.DataFrame(listPTA, columns=['Judul', 'NIM', 'Penulis', 'Abstrak'])
listPTA_df
```

```python
listPTA_df.to_csv('PTA_scraping.csv')
```
