```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
datas=[]
n = 5
url = 'https://news.detik.com/indeks/'
header={
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36'
}


for page in range(1,n+1):
  req = requests.get(url+str(page),headers=header)
  soup = BeautifulSoup(req.text, 'html.parser')
  berita = soup.findAll('article', 'list-content__item')
  # print(berita)
  for item in berita:
    link = item.find('a','media__link')['href']
    judul2 = item.find('h3','media__title').text.replace('\n','')
    print(judul2)
    req2 = requests.get(link)
    soup2 = BeautifulSoup(req2.text, 'html.parser')
    judul = soup2.find('h1', 'detail__title').text.replace('\n','')
    try:
      isi = soup2.find('div','detail__body-text itp_bodycontent').text.replace('\n','').replace('\t', '').replace('[Gambas:Video 20detik]', '')
    except:
      isi=''
    print(isi)
    datas.append([judul, isi])




```

```python
df = pd.DataFrame(datas, columns=['judul', 'isi'])
df
```

```python
df.to_csv('crawl_berita.csv')
```
